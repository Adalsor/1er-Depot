```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv("data.csv")
```

```python
df
```

```python
df["price_paid"] = df["price_paid"].apply(lambda x: x.replace("$", ""))
df["price_paid"] = df["price_paid"].astype(float)
```

```python
df
```

```python
df.groupby("date")["price_paid"].sum().plot()
```

```python
df.groupby("gender")["price_paid"].sum().plot.pie(legend=True, figsize=(6, 6))
```

```python
df.groupby("gender")["price_paid"].sum().plot.bar(legend=True, rot=45)
```
