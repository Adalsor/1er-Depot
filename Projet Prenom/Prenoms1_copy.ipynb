Quelle est la proportion totale Femmes / Hommes au cours du temps depuis 1900 ?

```python
nbr_annee = df.groupby(["annais", "sexe"])["nombre"].sum().reset_index()
```

```python
totaux_par_annee = nbr_annee.groupby("annais")["nombre"].sum().reset_index()
totaux_par_annee.rename(columns={"nombre": "total_naissances"}, inplace=True)
```

```python
nbr_annee = nbr_annee.merge(totaux_par_annee, on="annais")
```

```python
nbr_annee['nombre'] = nbr_annee['nombre'] / nbr_annee['total_naissances']
```

```python
nbr_annee
```

Elle est proche de 50/50
